````python
blinkit_df = pd.read_csv('https://raw.githubusercontent.com/Taj1920/Data-Analysis-Datasets/refs/heads/main/Datasets-Practice/blinkit_cleaned%20(1)')
blinkit_df
````

# 1. Plot a line graph showing the average Sales for each Outlet Establishment Year

````python
avg_sales_year = (
    blinkit_df.groupby('Outlet Establishment Year')['Sales']
      .mean()
      .sort_index()
)

plt.figure(figsize=(10,6))
plt.plot(avg_sales_year.index, avg_sales_year.values,marker='o',linestyle='-',color='skyblue',label='Average sales')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Average Sales')
plt.title('Average Sales by Outlet Establishment Year')
plt.xticks(avg_sales_year.index)
plt.annotate(f'Highest Sales({avg_sales_year.idxmax()}',xy=(avg_sales_year.idxmax(),avg_sales_year.max()),color='green')
plt.annotate(f'Lowest Sales({avg_sales_year.idxmin()}',xy=(avg_sales_year.idxmin(),avg_sales_year.min()),color='green')
plt.grid()
plt.legend()
plt.show()
````

# 2. Plot a line chart where the X-axis is Item Visibility (sorted) and the Y-axis is average Sales.


````python
avg_sales_by_visibility = blinkit_df.groupby('Item Visibility')['Sales'].mean().sort_index()
avg_sales_by_visibility
````

````python

plt.figure(figsize=(12, 6))
plt.plot(avg_sales_by_visibility.index, avg_sales_by_visibility.values, linestyle='-', color='blue', label='Average Sales')
plt.xlabel('Item Visibility', fontsize=12)
plt.ylabel('Average Sales', fontsize=12)
plt.title('Average Sales by Item Visibility', fontsize=14)
plt.legend()
plt.annotate(f'Highest Sales ({avg_sales_by_visibility.idxmax()}',xy=(avg_sales_by_visibility.idxmax(),avg_sales_by_visibility.max()),color='green')
plt.annotate(f'Lowest Sales ({avg_sales_by_visibility.idxmin()}',xy=(avg_sales_by_visibility.idxmin(),avg_sales_by_visibility.min()),color='red')
plt.grid()
plt.show()
````
